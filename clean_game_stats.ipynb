```python
# split rush
def split_rushing_stats(stat):
    split_stats = stat.split('-')
    att = split_stats[0]
    yds = split_stats[1]
    tds = split_stats[2]
    return att, yds, tds

def split_passing_stats(stat):
    split_stats = stat.split('-')
    # Cmp-Att-Yd-TD-INT	
    cmp = split_stats[0]
    att = split_stats[1]
    yds = split_stats[2]
    tds = split_stats[3]
    int = split_stats[4]
    return cmp, att, yds, tds, int

def split_sack_stats(stat):
    split_stats = stat.split('-')
    sacks = split_stats[0]
    yds = split_stats[1]
    return sacks, yds

def split_penalty_stats(stat):
    split_stats = stat.split('-')
    penalties = split_stats[0]
    yds = split_stats[1]
    return penalties, yds

rush_stats = split_rushing_stats('25-121-1')
pass_stats = split_passing_stats('26-31-297-3-2')
sack_stats = split_sack_stats('3-34')
penalty_stats = split_penalty_stats('2-30')
```

```python
def get_away_spread(spread_init, away_team, home_team):
    splits = str(spread_init).split(' ')
    spread_team = splits[-2].lower()
    spread = float(splits[-1])
    if away_team == spread_team:
        return spread
    elif home_team == spread_team:
        return spread * -1
    else:
        print(f'Error: couldnt match spread team \n Spread: {spread_init} \t Home: {home_team} \t Away: {away_team}')
    

get_away_spread('Indianapolis Colts -2.5', 'colts', 'chiefs')
```

```python
def clean_game_stats(wk_num):
    df = pd.read_csv(f'./game_stats/wk_{wk_num}_game_stats.csv')
    # remove columns
    df1 = df.drop(cols_to_remove, axis=1)
    # split rushing stats
    df1[['away_rush_att','away_rush_yds', 'away_rush_tds']] = df['rush_away'].str.split('-',expand=True)
    df1[['home_rush_att','home_rush_yds', 'home_rush_tds']] = df['rush_home'].str.split('-',expand=True)
    #split passing stats
    df1[['away_pass_cmp', 'away_pass_att','away_pass_yds', 'away_pass_tds', 'away_pass_ints']] = df['pass_away'].str.split('-',expand=True)
    df1[['home_pass_cmp', 'home_pass_att','home_pass_yds', 'home_pass_tds', 'home_pass_ints']] = df['pass_home'].str.split('-',expand=True)
    # split sacks stats
    df1[['away_sacks','away_sack_yds']] = df['sacked_away'].str.split('-', expand=True)
    df1[['home_sacks','home_sack_yds']] = df['sacked_home'].str.split('-', expand=True)
    # split penalyt stats
    df1[['home_penalty','home_penalty_yds']] = df['penalty_home'].str.split('-', expand=True)
    df1[['away_penalty','away_penalty_yds']] = df['penalty_away'].str.split('-', expand=True)
    # drop columns used to split
    df1 = df1.drop(['rush_away','rush_home', 'pass_away','pass_home','sacked_away','sacked_home', 'penalty_home','penalty_away'], axis=1)
    # convert spread from string to float
    # df1['away_spread'] = df['away_spread'].apply(lambda x: get_away_spread(x))

    df1['away_spread'] = df1.apply(
        lambda x: get_away_spread(
            x.away_spread, x.away, x.home
        ), 
        axis=1
    )
    # save to csv
    df1.to_csv(f'./game_stats/wk_{wk_num}_game_stats_clean2.csv', index=False)
    return df1
```
