```python
def find_row(team, week, df):
    index = df.loc[(df['week'] == week) & (df['team'] == team)].index
    if index.size == 0:
        return f'ERROR: could not find index for **{team}** and week **{week}**'
    return index

find_row('49ers', 1,  all_stats_df)
```

```python
def get_df_home(df, index):
    df = df.iloc[[index]]
    df = df.loc[:, keep_columns_game_home]
    df = df.rename(columns = rename_columns_game_home)
    df = df.reset_index(drop=True)
    df['spread'] = df['spread']*-1
    return df

def get_df_away(df, index):
    df = df.iloc[[index]]
    df = df.loc[:, keep_columns_game_away]
    df = df.rename(columns = rename_columns_game_away)
    df = df.reset_index(drop=True)
    return df

# for index, row in df_wk_1_small.iterrows():
#     df_home = get_df_home(df_wk_1_small, index)
#     # df_away = get_df_away(df_wk_1_small, index)
#     print(df_home)
```

```python


def get_game_stats_from_game(df_games, df_team_stats):
    lst = []
    columns = []
    for index, row in df_games.iterrows():
        print('Index: ', index)
        # get home and away team names and game week
        home = row['home']
        away = row['away']
        week = row['Week']
        print('HOME', home, '\t AWAY: ', away, '\t WEEK: ', week)
        
        # get home team stats from game stats
        df_home = get_df_home(df_games, index)

        # get away team stats from game stats
        df_away = get_df_away(df_games, index)

        # find index in all team stats based on team name and week number
        row_index_home = find_row(home, week, df_team_stats)
        row_index_away = find_row(away, week, df_team_stats)
        # print('\n Index Away: ', row_index_away[0], '\t Index Home: ', row_index_home[0], '\n -----------------\n\n')

        # merge stats together in new row
        home_team_row = pd.concat([df_team_stats.iloc[row_index_home].reset_index(drop=True), df_home], axis=1)
        away_team_row = pd.concat([df_team_stats.iloc[row_index_away].reset_index(drop=True), df_away], axis=1)

        # get values
        combo = pd.concat([home_team_row, away_team_row], axis=0)
        lst.append(combo)

        columns = combo.columns
    # create complete dataframe
    final_df = pd.concat(x for x in lst)
    final_df.columns = columns
    final_df = final_df.reset_index(drop=True)
    return final_df

final_df = get_game_stats_from_game(all_game_stats_df, all_stats_df)
```
